```python
from datetime import datetime
import numpy as np

# class LRUCacheItem(object):
#     """Data structure of items stored in cache"""
#     def __init__(self, item):
# #         self.key = key
#         self.item = item
# #         self.timestamp = datetime.now()


class LRUCache(object):
    """A sample class that implements LRU algorithm"""

    def __init__(self, length, delta=None):
        self.length = length
#         self.delta = delta
#         self.hash = {}
        self.item_list = []
        self.hash_set = set()
#         self.arr = np.array([])
        
                    #     def insertItemArr(self, item):
                    #         """Insert new items to cache"""

                    #         cache_hit = False
                    #         if item in self.hash_set:
                    #             self.arr = self.arr[self.arr != item]
                    #             np.concatenate([[item], self.arr])
                    #         else:
                    #             # Remove the last item if the length of cache exceeds the upper bound.
                    #             if self.arr.size >= self.length:
                    #                 self.removeItemArr(self.arr)

                    #             # If this is a new item, just append it to
                    #             # the front of item_list.
                    #             self.hash_set.add(item)
                    #             np.concatenate([[item], self.arr])

                    #         print ("self.hash_set = ",self.hash_set)
                    #         print ("self.arr = ", self.arr)
                    #         return cache_hit                

                    #     def removeItemArr(self, item):
                    #         """Remove those invalid items"""

                    #         self.hash_set.remove(item)
                    #         self.arr[np.arange(self.arr.size - 1)]
            
        

    def insertItem(self, item):
        """Insert new items to cache"""
        cache_hit = False
        if item in self.hash_set:
            item_index = self.item_list.index(item)
            del self.item_list[item_index]
            self.item_list.insert(0, item)
            cache_hit = True
        else:
            if len(self.item_list) >= self.length:
                self.removeItem(self.item_list[-1])

            self.hash_set.add(item)
            self.item_list.insert(0, item)
        for i in self.item_list:
            print ("{}  |".format(i), end=" ")
        return cache_hit

    def removeItem(self, item):
        """Remove those invalid items"""

        self.hash_set.remove(item)
        self.item_list.pop()
```
